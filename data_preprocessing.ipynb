### Import necessary library's

```python
import json
import pandas as pd
```

### Transform the file from jso to csv and read it after that apply somme function to clean the weird caracteres

```python
with open('./data_json_files/Match_results.json') as f:
    data =json.load(f)
match_data = pd.DataFrame(data[0]['matchSummary'])
match_data.head(5)
```

```python
#rename the column scorecard by match_id
match_data.rename({'scorecard':'match_id'},axis=1,inplace=True)
```

```python
#save the file
match_data.to_csv('./data_csv_files/Match_results.csv',index=False)
```

### Transform the batting summary file from json to csv and read it after that apply somme function to clean the weird caracteres

```python
with open('./data_json_files/Batting_summary.json') as f:
    data = json.load(f)

all_batting = []
for batting in data:
    all_batting.extend(batting['battingSummary'])

batting_summary_data = pd.DataFrame(all_batting)
batting_summary_data.tail(5)
```

##### add new column out/not out using dismissal column

```python
batting_summary_data['out/not out'] = batting_summary_data['dismissal'].apply(lambda x: 'out' if len(x)>0 else 'not out')
batting_summary_data.head(5)
```

##### drop dismissal column from dataframe 

```python
batting_summary_data.drop(columns={'dismissal'},inplace=True)
batting_summary_data.head(5)
```

##### drop somme weird caracteres

```python
batting_summary_data['batsmanName']=batting_summary_data['batsmanName'].apply(lambda x: x.replace('â€',''))
batting_summary_data['batsmanName']=batting_summary_data['batsmanName'].apply(lambda x: x.replace('\xa0',''))
```

```python
batting_summary_data[batting_summary_data['match']=='Namibia Vs Sri Lanka']
```

#### create a dictionairy to help us to Joining two tables

```python
dic_ids_match = {}
for ids,rows in match_data.iterrows():
    key1 = rows['team1'] + " Vs " + rows['team2']
    key2 = rows['match_id']
    dic_ids_match[key1] = key2


first_items = dict(list(dic_ids_match.items())[:5])
print(first_items)
```

```python
batting_summary_data['match_id'] = batting_summary_data['match'].map(dic_ids_match)
batting_summary_data.head(4)
```

```python
#save the file
batting_summary_data.to_csv('./data_csv_files/Batting_summary.csv',index=False)
```

### Transform the Bowling summary file from json to csv and read it after that apply somme function to clean the weird caracteres

```python
with open('./data_json_files/Bowling_summary.json') as f:
    data = json.load(f)
all_bowling = []
for bowl in data:
    all_bowling.extend(bowl['bowlingSummary'])

all_bowling_summary = pd.DataFrame(all_bowling)
all_bowling_summary.head(5)
```
